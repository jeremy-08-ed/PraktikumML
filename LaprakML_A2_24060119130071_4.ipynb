##Import Library

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd
```

##Import Dataset

```python
# membuat data set
from sklearn.datasets import load_iris 
iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

##Plotting Kolom Indeks 2 dan 3

```python
# plotting
features=iris.data.T
plt.plot()
plt.scatter(features[2], features[3]) 
plt.show()
```

Sangat terlihat apabila dalam plot tersebut, terdapat 2 grup yang berkelompok.

##Menentukan Nilai Sum Squared Errors (SSE) pada setiap K

```python
X = np.array(list(zip(features[2], features[3]))).reshape(len(features[3]), 2)
SSE = []
K = range(1, 10)
for k in K:
  kmeanModel = KMeans(n_clusters=k).fit(X)
  kmeanModel.fit(X)
  SSE.append(kmeanModel.inertia_)

print(SSE)
```

##Plotting Nilai SSE dengan $1 \le K \le 9$

```python
plt.plot(K, SSE, 'co-')
plt.xlabel('k')
plt.ylabel('SSE')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

'co-' merupakan warna cyan dengan bentuk lingkaran

Nilai K ditentukan 2 karena ketika K = 1, grafik sangat curam dan ketika K = 2, grafik sudah lebih landai (tidak securam ketika K = 1 ke K = 2)

```python
y_pred = KMeans(n_clusters=2).fit_predict(X)
plt.plot
LABEL_COLOR_MAP = {0: 'r',
                   1: 'g',
                   2: 'b'
                   }
label_color = [LABEL_COLOR_MAP[l] for l in y_pred]
plt.scatter(features[2], features[3], c=label_color)
plt.title("K=2")
plt.show()
```

##Clustering dengan $ 1 \le K \le 6$

###`cmap='coolwarm'`

```python
figure,ax = plt.subplots(2,3)
K = range(1,7)
for k in K:
  if(k < 4):
    row = 0
    column = k - 1
  else:
    row = 1
    column = k - 4
  kmeanModel = KMeans(n_clusters=k).fit(X)
  y_pred = kmeanModel.fit_predict(X)
  ax[row][column].scatter(features[2], features[3], c=y_pred, cmap='coolwarm')

plt.show()
```
